```python
#Thống kê dữ liệu thiếu cho cột dân tộc và hiệu chỉnh dữ liệu thiếu như sau: Mặc định thiếu thì điền giá trị 0. 
df["DT"].fillna(0, inplace=False)
```

```python
#Thống kê dữ liệu thiếu cho biến T1 và hiệu chỉnh dữ liệu, lưu ý việc thay thế dữ liệu thiếu sử dụng phương pháp Mean.
df["DT"].fillna(df["T1"].mean(), inplace=False)
```

```python
def fill(col):
  return df[col].fillna(df[col].mean(), inplace =False)
```

```python
is_float64 = df.dtypes == float
df_column_list = df.columns[is_float64].to_list()
```

```python
df_column_list
```

```python
for col in df_column_list:
  df[col].fillna(df[col].mean(), inplace =False)
```

```python
for i in ["1","2","6"]:
  print("TBM"+i)
```

```python
for i in ["1","2","3"]:
  df["TBM"+i] = (df["T"+i]*2 + df["L"+i] + df["H"+i]+ df["S"+i]+ + df["V"+i]*2 +df["X"+i] +df["D"+i]+ df["N"+i])/10
```

```python
df.rename(columns= {"TBM":"TBM3"}, inplace= False)
```

```python
df.head()
```

```python

#8
df['XL1'] = df['TBM1'].apply(lambda x: 'Y' if x < 5.0 else ('TB' if x < 6.5 else ('K' if x < 8.0 else ('G' if x < 9.0 else 'XS'))))
print(df)
```

```python

df['XL2'] = df['TBM2'].apply(lambda x: 'Y' if x < 5.0 else ('TB' if x < 6.5 else ('K' if x < 8.0 else ('G' if x < 9.0 else 'XS'))))

# Hiển thị DataFrame sau khi thêm các biến xếp loại
print(df)
```

```python
df['XL3'] = df['TBM3'].apply(lambda x: 'Y' if x < 5.0 else ('TB' if x < 6.5 else ('K' if x < 8.0 else ('G' if x < 9.0 else 'XS'))))

# Hiển thị DataFrame sau khi thêm các biến xếp loại
print(df)
```

```python
#9
df['US_TBM1'] = df['TBM1'] / 10 * 4

print(df)
```

```python

df['US_TBM2'] = df['TBM2'] / 10 * 4

print(df)
```

```python
df['US_TBM3'] = df['TBM3'] / 10 * 4
print(df)
```

```python
#10
# Tạo biến KQXT cho khối A và A1
df['KQXT_A'] = df.apply(lambda row: 1 if ((row['DH1'] * 2 + row['DH2'] + row['DH3']) / 4) >= 5.0 else 0, axis=1)

# Tạo biến KQXT cho khối B
df['KQXT_B'] = df.apply(lambda row: 1 if ((row['DH1'] + row['DH2'] * 2 + row['DH3']) / 4) >= 5.0 else 0, axis=1)

# Tạo biến KQXT cho khối khác
df['KQXT_K'] = df.apply(lambda row: 1 if ((row['DH1'] + row['DH2'] + row['DH3']) / 3) >= 5.0 else 0, axis=1)

# Hiển thị DataFrame sau khi thêm các biến KQXT
print(df)
```

```python
#11
df.to_csv('processed_dulieuxettuyendaihoc.csv', index=False)
```
