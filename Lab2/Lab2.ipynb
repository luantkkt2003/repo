2. Hãy tạo biến phân loại (phanlopt1) cho môn toán (T1) như sau: 
a. Từ 0 đến dưới 5 = kém (ký hiệu k)  
b. Từ 5 đến dưới 7 = trung bình (ký hiệu tb) 
c. Từ 7 đến dưới 8 = khá (ký hiệu k) 
d. Từ 8 trở lên = giỏi (ký hiệu g)

```python
df.loc[(df['T1'] < 5), 'phanlopt1']= 'kem'
df.loc[(df['T1'] >= 5) & (df['T1'] < 7), 'phanlopt1']= 'trungbinh'
df.loc[(df['T1'] >= 7) & (df['T1'] < 8), 'phanlopt1']= 'kha'
df.loc[(df['T1'] >= 8), 'phanlopt1']= 'gioi'
#a. Từ 0 đến dưới 5 = kém (ký hiệu k)  
#b. Từ 5 đến dưới 7 = trung bình (ký hiệu tb) 
#c. Từ 7 đến dưới 8 = khá (ký hiệu k) 
#d. Từ 8 trở lên = giỏi (ký hiệu g)
```

4. Vẽ biểu đồ đường Multiple Line cho biến T1 được phân loại bởi biến phanlopt1 

```python
import pandas as pd
import matplotlib.pyplot as plt
grouped_data = df.groupby('phanlopt1')['T1']
#Nhóm cột phanlopt1 theo giá trị T1 

# Vẽ biểu đồ đường Multiple Line
plt.figure(figsize=(10, 6))
#vẽ biểu đồ dài 10 ngang 6
for name, group in grouped_data:
    group.plot(label=name)

plt.title('Biểu đồ đường Multiple Line của T1 theo phanlopt1')
plt.xlabel('STT')
plt.ylabel('Giá trị của T1')
plt.legend()
plt.show()
```

5. Vẽ biểu đồ Drop-line cho biến T1 được phân loại bởi biến phanlopt1

```python

import pandas as pd
import matplotlib.pyplot as plt
# Nhóm dữ liệu dựa trên biến "phanlopt1" và tính giá trị trung bình của T1
grouped_data = df.groupby('phanlopt1')['T1'].mean()

# Vẽ biểu đồ Drop-line
plt.figure(figsize=(10, 6))
for i, (name, value) in enumerate(grouped_data.items()):
    plt.plot([i, i], [0, value], marker='o', linestyle='-', label=name)

plt.xticks(range(len(grouped_data)), grouped_data.index)
plt.title('Biểu đồ Drop-line của T1 theo phanlopt1')
plt.xlabel('Phân loại')
plt.ylabel('Giá trị trung bình của T1')
plt.legend()
plt.show()
```
